# 1. Présentation du projet

Ce projet est réalisé dans le cadre du défi **Customer NPS Prediction for a Telecom Operator**, qui simule une problématique rencontrée par un opérateur télécom panafricain. L'objectif est de développer un système de machine learning capable de prédire la catégorie **Net Promoter Score (NPS)** d'un client, à savoir **Détracteur**, **Passif** ou **Promoteur**, à partir de ses informations de compte, des services auxquels il est abonné et de son historique d'interactions avec l'entreprise.

Dans de nombreux opérateurs télécoms, seule une faible proportion des clients répond aux enquêtes de satisfaction. Il devient alors difficile pour les équipes chargées de la fidélisation d'identifier les clients insatisfaits et d'intervenir avant qu'ils ne quittent l'entreprise. En prédisant le NPS des clients qui ne répondent pas aux enquêtes, ce projet vise à fournir un outil d'aide à la décision permettant de cibler plus efficacement les actions de rétention.

Pour mener cette étude, nous utiliserons le jeu de données public **IBM Telco Customer Churn**, qui regroupe les informations d'environ 7 000 clients d'un opérateur télécom fictif. Ce jeu de données contient notamment un **Satisfaction Score** compris entre 1 et 5, qui servira de point de départ pour construire la variable cible représentant les catégories du NPS.

Au-delà de la construction d'un modèle prédictif, ce projet suit une démarche complète de data science. Il couvre l'ensemble des étapes nécessaires, depuis la compréhension du besoin métier jusqu'au déploiement d'un prototype fonctionnel. Les différentes phases comprennent la construction de la variable cible, la préparation des données, l'ingénierie des caractéristiques, l'entraînement et l'évaluation des modèles, l'interprétation des résultats, l'analyse des facteurs expliquant l'insatisfaction des clients, l'audit des biais potentiels, puis le développement d'une interface permettant aux équipes métier d'utiliser facilement les prédictions obtenues.

# 2. Contexte métier

L’opérateur télécom réalise régulièrement des enquêtes Net Promoter Score (NPS) afin de mesurer la satisfaction et la fidélité de ses clients. Dans le cadre de ce projet, environ **15 %** des clients fournissent une réponse NPS.

Cela crée une limite importante pour l’entreprise : elle ne dispose d’un retour direct que pour une faible partie de sa base client. Les **85 % restants** ne s’expriment pas, ce qui rend difficile pour les équipes marketing et expérience client de comprendre leur niveau de satisfaction et de décider quels clients doivent être contactés en priorité.

Il s’agit d’un enjeu business important. Dans le secteur des télécoms, retenir un client coûte généralement beaucoup moins cher que d’en acquérir un nouveau. Si les clients insatisfaits ne sont pas identifiés à temps, ils risquent de partir chez un concurrent, ce qui entraîne une perte de revenus et peut aussi nuire à la réputation de l’entreprise à travers le bouche-à-oreille négatif ou les avis en ligne.

C’est pourquoi l’entreprise souhaite utiliser le machine learning pour estimer la catégorie NPS des clients qui ne répondent pas aux enquêtes. Ces prédictions permettront à l’équipe de rétention d’identifier de manière proactive les clients susceptibles d’être des **Détracteurs**, de prioriser les actions de contact et de prendre de meilleures décisions à grande échelle.


# 3. Problème métier

L'entreprise ne peut pas mettre en place une stratégie de rétention efficace lorsqu'elle ne connaît le niveau de satisfaction que de **15 %** de ses clients.

Cette situation entraîne plusieurs difficultés. Tout d'abord, les équipes de rétention ne savent pas si la majorité des clients sont satisfaits, neutres ou potentiellement insatisfaits. Elles interviennent donc souvent après une résiliation ou une réclamation, plutôt que d'agir en amont pour éviter la perte du client.

Par ailleurs, l'entreprise peine à comprendre les principaux facteurs qui influencent la satisfaction de ses clients. Sans cette visibilité, il devient difficile d'identifier les leviers d'amélioration, qu'il s'agisse de la qualité des services, des contrats proposés ou de l'expérience client.

Enfin, les décisions prises à partir des seules réponses aux enquêtes NPS peuvent être biaisées. Les clients qui répondent ne sont pas nécessairement représentatifs de l'ensemble de la clientèle. L'entreprise risque donc d'orienter ses actions de rétention et ses ressources sur la base d'une vision incomplète de la satisfaction de ses clients.

# 4. Objectif métier

Construire un système de machine learning capable de prédire la catégorie NPS d’un client (Détracteur, Passif ou Promoteur) à partir de ses données de compte et de comportement, afin de rendre la satisfaction client exploitable à grande échelle.

Ce système permettra à l’équipe de rétention de :

* prioriser les actions de contact proactif vers les clients identifiés comme Détracteurs, avant qu’ils ne résilient ;
* comprendre les principaux facteurs de détraction, aussi bien au niveau individuel (pour adapter les actions de manière personnalisée) qu’au niveau segment (pour orienter les décisions produit, tarifaires ou d’expérience client) ;
* estimer le niveau de satisfaction des 85 % de clients qui ne répondent pas aux enquêtes, afin d’obtenir une vision plus complète et moins biaisée de la base client.


# 5. Objectif Machine Learning

Du point de vue de la data science, ce projet consiste à développer un modèle de machine learning capable de prédire la catégorie NPS d'un client (**Détracteur**, **Passif** ou **Promoteur**) à partir de ses données de compte, de ses services, de son comportement et, éventuellement, de données textuelles issues de ses dernières interactions avec le service client.

Il s'agit d'un problème de **classification supervisée**, puisque la variable cible est connue pour une partie des clients et que le modèle apprend à prédire cette catégorie à partir de variables explicatives.

Une particularité importante de ce projet est que les catégories NPS possèdent un ordre naturel : un **Détracteur** est plus proche d'un **Passif** qu'un **Promoteur**. Cette propriété fait du NPS une **variable ordinale**, ce qui ouvre plusieurs possibilités de modélisation. Selon les résultats obtenus, le problème pourra être abordé comme une classification multi-classe classique, une classification ordinale ou encore une régression suivie d'un découpage en catégories.

Le choix de l'approche retenue sera justifié lors de la phase de modélisation, en tenant compte des performances obtenues ainsi que des contraintes du problème métier.


# 6. Variable cible

La variable cible n’est pas directement disponible dans le jeu de données sous la forme d’un NPS à trois classes. Elle doit être construite à partir du **Score de Satisfaction**, fourni par le dataset IBM Telco, et qui est compris entre 1 et 5.

Pour transformer cette variable en catégories NPS exploitables par le modèle, un mapping de référence est utilisé comme point de départ :

| Score de satisfaction | Catégorie NPS | Équivalent NPS |
| --------------------- | ------------- | -------------- |
| 5                     | Promoteur     | 9 à 10         |
| 4                     | Passif        | 7 à 8          |
| ≤ 3                   | Détracteur    | 0 à 6          |


Ce mapping constitue une hypothèse de travail. Il pourra être ajusté ou enrichi en fonction de la cohérence des données et de l’impact observé sur la performance du modèle, notamment pour limiter les biais ou introduire plus de réalisme dans la distribution des classes.

## Point de vigilance : risque de data leakage
Le jeu de données contient également des variables telles que **Churn Score** et **Churn Value**, qui sont calculées par IBM à partir d’informations potentiellement liées à l’attrition des clients.

# 7. Critères de succès

## Critères de succès

L’évaluation du projet repose à la fois sur des critères métier et des critères techniques, afin de s’assurer que le modèle est non seulement performant, mais également utile dans un contexte opérationnel.

### Succès métier

Le projet sera considéré comme réussi si le modèle permet :

* d’identifier de manière fiable une part significative des clients Détracteurs, avec un rappel élevé sur cette classe, afin de rendre les campagnes de rétention plus efficaces ;
* de produire une estimation exploitable du niveau de satisfaction pour les clients qui ne répondent pas aux enquêtes, représentant environ 85 % de la base ;
* de mettre en évidence des facteurs de détraction compréhensibles et actionnables, en distinguant clairement les leviers sur lesquels l’entreprise peut agir (contrat, tarification, qualité de service) des variables non modifiables ;
* de fournir un résultat facilement interprétable et utilisable par des équipes non techniques, notamment les responsables de l’expérience client.

### Succès technique

Sur le plan technique, le projet doit respecter plusieurs exigences :

* la mise en place d’une baseline simple et robuste avant toute approche avancée de modélisation ;
* l’utilisation de métriques adaptées à un problème de classification ordinale et déséquilibrée, telles que le F1 macro, l’accuracy équilibrée, le kappa pondéré quadratique, le rappel par classe, ainsi que des analyses de calibration et de lift sur la classe Détracteur ;
* une stratégie de validation cohérente avec le contexte réel d’utilisation, en simulant la prédiction sur des clients n’ayant pas répondu aux enquêtes ;
* une attention particulière portée à l’équité du modèle, en vérifiant l’absence de disparités significatives de performance entre segments démographiques ;
* la mise à disposition d’une interface fonctionnelle permettant à un utilisateur métier d’obtenir une prédiction accompagnée de ses principaux facteurs explicatifs.


# 8. Contraintes et risques

### Contraintes

Le projet est limité par plusieurs contraintes assez concrètes.

D’abord, on a un délai de **deux semaines**, donc il faut rester focus sur l’essentiel et éviter de complexifier inutilement le pipeline.

Ensuite, on doit utiliser principalement le dataset **IBM Telco Customer Churn (11.1.3+)**. On peut ajouter des données externes, mais seulement si ça a un vrai intérêt pour le problème et qu’on peut le justifier.

Enfin, si on utilise des outils externes comme des LLM pour générer du texte, il faut faire attention à ne pas exposer de **clés API** ou d’informations sensibles dans le code.



### Risques

Il y a plusieurs risques importants à prendre en compte.

Le premier, c’est le **data leakage**. Des variables comme *Churn Score* ou *Churn Value* peuvent contenir des infos trop proches de la cible. Si on les utilise mal, le modèle peut sembler très performant alors qu’en réalité il ne serait **pas fiable en production**.

Il y a aussi le fait que la cible NPS n’est pas directement donnée. Elle est construite à partir d’un score de satisfaction, donc il y a forcément une part d’**hypothèse** dans le label.

Un autre problème, c’est le **déséquilibre des classes**. En général, les Détracteurs sont moins nombreux, donc le modèle peut avoir tendance à les rater, alors que ce sont justement les plus importants à détecter.

Il faut aussi faire attention aux **biais**. Certaines variables comme l’âge ou la zone géographique peuvent influencer indirectement les décisions et créer des différences de traitement entre groupes de clients.

Enfin, il y a un problème de **generalization shift**. Le modèle est entraîné sur les clients qui ont répondu aux enquêtes, mais il est utilisé sur ceux qui n’ont pas répondu. Ce n’est pas forcément la même population, donc les performances peuvent changer en vrai.

Et il faut garder en tête une chose simple : ce que le modèle apprend, ce sont des **corrélations**, pas des causes.


# 9. Hypothèses

Le projet repose sur plusieurs hypothèses importantes qu’il faut garder en tête lors de l’analyse et de la modélisation.

Le **Score de Satisfaction** fourni par les 15 % de répondants est considéré comme un proxy raisonnable, bien qu’imparfait, de leur intention de recommandation (NPS). Il permet de construire une cible exploitable, mais il ne reflète pas parfaitement toutes les nuances de la satisfaction client.

On suppose également que les clients répondants ne sont pas totalement différents des non-répondants. Même s’il existe probablement un biais de sélection, on considère qu’ils partagent suffisamment de caractéristiques pour permettre au modèle de généraliser, à condition de valider correctement les performances.

Les variables structurées disponibles (ancienneté, type de contrat, facturation, services souscrits, comportement client) sont supposées contenir un signal suffisant pour prédire la catégorie NPS, même sans informations textuelles.

L’ajout de verbatims clients générés par LLM peut apporter un signal supplémentaire intéressant, mais leur nature synthétique impose de rester prudent sur leur fidélité au langage réel des clients.

Le jeu de données IBM Telco, bien qu’artificiel, est considéré comme suffisamment réaliste pour permettre d’extraire des tendances exploitables sur les facteurs de détraction et les comportements clients.

Enfin, on suppose qu’une amélioration du rappel sur la classe Détracteur peut se traduire en valeur métier, notamment via une meilleure allocation des actions de rétention, même si cette relation n’est pas directement mesurée dans ce projet.


# 10. Questions d'analyse

Plusieurs questions doivent guider l’exploration des données et la construction du modèle.

On commence par s’intéresser à la distribution de la cible : quelle est la répartition du Score de Satisfaction (et donc des catégories NPS dérivées) parmi les répondants ? Le problème est-il fortement déséquilibré, et si oui, dans quelle mesure cela influence-t-il le choix des métriques et des modèles ?

Un point important concerne les variables *Churn Score* et *Churn Value*. Il faut vérifier dans quelle mesure elles sont corrélées avec le Score de Satisfaction, afin d’évaluer le risque de fuite de données si elles étaient utilisées dans le modèle.

Il est également essentiel de comprendre s’il existe un biais de sélection entre les clients répondants et non-répondants. Par exemple, observe-t-on des différences en termes d’ancienneté, de type de contrat ou de profil client ? Ce point est critique pour la capacité de généralisation du modèle.

Ensuite, il faut identifier quelles caractéristiques (contrat, facturation, services, ancienneté) sont les plus associées à chaque catégorie NPS, afin de mieux comprendre les drivers de satisfaction et d’insatisfaction.

Une autre question importante est de savoir si la relation entre les variables et la cible est suffisamment ordinale pour justifier une approche de classification ordinale, plutôt qu’un problème de classification multi-classe classique.

On s’intéressera aussi aux segments de clients les plus à risque : quels profils (nouveaux clients, contrats mensuels, type de service, etc.) présentent les niveaux de détraction les plus élevés ?

Dans une logique d’équité, il faudra vérifier si le modèle présente des différences de performance entre groupes démographiques ou géographiques, notamment en termes de rappel sur la classe Détracteur.

# 11. Conclusion

## Conclusion de la phase Business Understanding

Cette phase de Business Understanding pose les bases du projet. L’entreprise fait face à un manque important de visibilité sur la satisfaction de ses clients, puisque seulement **15 %** de la base est couverte par une mesure directe du NPS.

L’enjeu principal est double : d’un côté, mieux identifier les clients à risque afin de prioriser les actions de rétention, et de l’autre, étendre cette compréhension de la satisfaction à l’ensemble de la base client grâce à un modèle de prédiction.

Traduire cet enjeu en problème de machine learning implique déjà plusieurs décisions importantes. Il faut prendre en compte la nature **ordinale** de la cible, le risque de **data leakage** lié à certaines variables comme le Churn Score et la Churn Value, ainsi que le déséquilibre des classes. Il faut également anticiper les questions de **généralisation** et d’**équité entre les segments de clients**.

Ces éléments vont guider toutes les étapes suivantes du projet, de la préparation des données jusqu’à la modélisation et l’évaluation.
