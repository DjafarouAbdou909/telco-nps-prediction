# 1. Project Overview

This project is part of the **Customer NPS Prediction for a Telecom Operator** challenge, which simulates a real business problem faced by a pan-African telecom company. The goal is to build a machine learning system that can predict a customer's **Net Promoter Score (NPS)** category: **Detractor**, **Passive**, or **Promoter**, using customer account information, subscribed services, and interaction history.

In many telecom companies, only a small number of customers answer satisfaction surveys. Because of this, customer retention teams have limited information to identify dissatisfied customers and take action before they leave the company. By predicting the NPS of customers who do not answer these surveys, this project aims to provide a decision support tool that helps the company target its retention efforts more effectively.

To carry out this study, we will use the public **IBM Telco Customer Churn** dataset, which contains information about nearly **7,000** customers from a fictional telecom company. The dataset includes a **Satisfaction Score** ranging from **1 to 5**, which will be used to create the target variable representing the three NPS categories.

Beyond building a predictive model, this project follows a complete data science workflow. It covers every major step, from understanding the business problem to developing a working prototype. The project includes target variable creation, data preparation, feature engineering, model training and evaluation, result interpretation, analysis of the main factors behind customer dissatisfaction, bias assessment, and the development of a simple interface that allows business users to easily use the model's predictions.


## 2. Business Context

The telecom operator regularly conducts Net Promoter Score (NPS) surveys to measure customer satisfaction and loyalty. In this project, about **15% of customers** provide an NPS response.

This creates an important limitation for the company. It only receives direct feedback from a small part of its customer base. The remaining **85% of customers** do not respond, which makes it difficult for marketing and customer experience teams to understand their satisfaction level and decide which customers should be contacted first.

This is a key business problem. In the telecom industry, keeping an existing customer is usually much cheaper than acquiring a new one. If dissatisfied customers are not identified early, they may leave for competitors, which leads to revenue loss and can also harm the company’s reputation through negative word of mouth or online reviews.

For this reason, the company wants to use machine learning to estimate the NPS category of customers who do not respond to surveys. These predictions will help the retention team identify likely **Detractors**, prioritize actions, and improve decision-making at scale.


## 3. Business Problem

The company cannot build an effective customer retention strategy when it only knows the satisfaction level of **15% of its customers**.

This creates several challenges. First, the retention team does not know whether most customers are satisfied, neutral, or dissatisfied. As a result, they often take action only after a customer has left the company or made a complaint, instead of acting early to prevent customer loss.

The company also has difficulty understanding the main factors that influence customer satisfaction. Without this information, it is hard to identify what should be improved, whether it is service quality, contract offers, or the overall customer experience.

Finally, decisions based only on NPS survey responses may be biased. Customers who answer the survey are not always representative of the entire customer base. As a result, the company may focus its retention efforts and resources on an incomplete view of customer satisfaction.

## 4. Business Objective

The goal of this project is to build a machine learning system that can predict a customer's NPS category (**Detractor**, **Passive**, or **Promoter**) using customer account and behavioral data. The aim is to make customer satisfaction measurable for the entire customer base.

This system will help the retention team:

* prioritize proactive actions for customers predicted to be **Detractors**, before they leave the company;
* understand the main factors behind customer dissatisfaction, both at the individual level (to personalize retention actions) and at the customer segment level (to support product, pricing, and customer experience decisions);
* estimate the satisfaction level of the **85% of customers** who do not respond to NPS surveys, giving the company a more complete and less biased view of customer satisfaction.


## 5. Machine Learning Objective

From a data science perspective, this project aims to build a machine learning model that can predict a customer's NPS category (**Detractor**, **Passive**, or **Promoter**) using customer account information, subscribed services, customer behavior, and, if available, text data from recent customer support interactions.

This is a **supervised classification** problem because the target variable is known for a group of customers, and the model learns to predict this target from the available features.

One important aspect of this project is that the NPS categories have a natural order. A **Detractor** is closer to a **Passive** than to a **Promoter**. This means that NPS is an **ordinal variable**, which allows different modeling approaches. Depending on the results, the problem can be solved using a standard multi-class classification model, an ordinal classification model, or a regression model followed by a conversion into NPS categories.

The final modeling approach will be chosen during the modeling phase based on the model's performance and the business requirements.


## 6. Target Variable

The target variable is not directly available in the dataset as a three-class NPS label. It must be created from the **Satisfaction Score** provided in the IBM Telco dataset, which ranges from **1 to 5**.

To convert this score into NPS categories, we use the following mapping as a starting point:

| Satisfaction Score | NPS Category | Equivalent NPS |
| ------------------ | ------------ | -------------- |
| 5                  | Promoter     | 9–10           |
| 4                  | Passive      | 7–8            |
| ≤ 3                | Detractor    | 0–6            |

This mapping is a **working assumption**. It may be adjusted later if needed, depending on the data analysis and its impact on the model's performance.

### Important Note: Data Leakage Risk

The dataset also contains variables such as **Churn Score** and **Churn Value**, which were calculated by IBM using information that may be related to customer churn.


## 7. Success Criteria

The evaluation of this project is based on both business and technical criteria, ensuring that the model is not only accurate, but also useful in a real operational context.

### Business Success

The project will be considered successful if the model can:

* reliably identify a significant portion of **Detractor** customers, with high recall on this class, in order to make retention campaigns more effective;
* provide a usable estimate of customer satisfaction for the **85% of customers** who do not respond to NPS surveys;
* highlight clear and actionable drivers of customer dissatisfaction, distinguishing between factors the company can influence (contracts, pricing, service quality) and those that cannot be changed;
* produce results that are easy to interpret and usable by non-technical teams, especially customer experience managers.

### Technical Success

From a technical perspective, the project must satisfy several requirements:

* implementation of a simple and robust **baseline model** before moving to more advanced approaches;
* use of appropriate metrics for an ordinal and imbalanced classification problem, such as macro F1-score, balanced accuracy, quadratic weighted kappa, class-wise recall, as well as calibration and lift analysis for the Detractor class;
* a validation strategy that reflects real-world usage by simulating predictions on customers who did not respond to surveys;
* careful evaluation of model fairness, ensuring no significant performance gaps across demographic groups;
* delivery of a functional interface that allows business users to generate predictions along with their main explanatory factors.

## 8. Constraints and Risks

### Constraints

The project is limited by several practical constraints.

First, there is a **two-week deadline**, which means the focus must stay on the most important steps and avoid unnecessary complexity in the pipeline.

Second, the project mainly uses the **IBM Telco Customer Churn (11.1.3+) dataset**. External data can be added, but only if it brings clear value to the problem and is properly justified.

Finally, if external tools such as LLMs are used to generate text, special attention must be given to security, especially to avoid exposing **API keys or sensitive information** in the codebase.



### Risks

Several important risks must be considered.

The first one is **data leakage**. Variables such as *Churn Score* and *Churn Value* may contain information too closely related to the target. If used incorrectly, the model may show strong performance during testing but fail in real production use.

Another risk is related to the target itself. The NPS label is not directly observed in the dataset but constructed from a satisfaction score. This introduces a level of **assumption and uncertainty** in the ground truth.

There is also a **class imbalance problem**, where Detractors are usually less frequent. This can lead the model to perform poorly on the most important class from a business perspective.

**Bias and fairness** are also concerns. Some features such as age or geographic location may indirectly influence predictions and lead to unequal treatment across customer groups.

In addition, there is a **generalization shift risk**. The model is trained on customers who responded to surveys but is applied to those who did not. These two populations may not follow the same distribution, which can affect real-world performance.

Finally, it is important to remember that the model learns **correlations, not causal relationships**. Therefore, any business interpretation of the results must be done carefully.

## 9. Assumptions


This project is based on several important assumptions that should be kept in mind during analysis and modeling.

The **Satisfaction Score** provided by the 15% of respondents is considered a reasonable but imperfect proxy for their Net Promoter Score (NPS). It allows us to build a usable target variable, but it does not fully capture all aspects of customer satisfaction.

We also assume that respondents are not completely different from non-respondents. Even though there may be some selection bias, we consider that they share enough similar characteristics for the model to generalize, provided that proper validation is performed.

The available structured features (such as tenure, contract type, billing information, subscribed services, and customer behavior) are assumed to contain enough signal to predict the NPS category, even without text data.

The use of LLM-generated customer verbatims may provide additional useful information, but their synthetic nature requires caution regarding how well they represent real customer language.

Although the IBM Telco dataset is synthetic, it is assumed to be realistic enough to capture meaningful patterns in customer behavior and dissatisfaction drivers.

Finally, we assume that improving recall on the Detractor class can translate into business value, especially through better allocation of retention actions, even if this causal relationship is not directly measured in this project.



## 10. Analysis Questions

Several questions should guide the data exploration and model building process.

We first focus on the target distribution: what is the distribution of the **Satisfaction Score** (and therefore the derived NPS categories) among respondents? Is the problem highly imbalanced, and if so, how does this influence the choice of metrics and models?

An important point concerns the **Churn Score** and **Churn Value** variables. It is necessary to check how strongly they are correlated with the Satisfaction Score in order to assess the risk of data leakage if they are used in the model.

It is also important to understand whether there is a selection bias between respondents and non-respondents. For example, are there differences in tenure, contract type, or customer profile? This is critical for the model’s ability to generalize.

Next, we need to identify which features (contract, billing, services, tenure) are most strongly associated with each NPS category, in order to better understand the drivers of satisfaction and dissatisfaction.

Another key question is whether the relationship between features and the target is sufficiently ordinal to justify an ordinal classification approach instead of a standard multi-class classification problem.

We also need to analyze which customer segments are most at risk. For example, which profiles (new customers, monthly contracts, specific services, etc.) show the highest levels of detractors?

Finally, from a fairness perspective, we should check whether the model shows performance differences across demographic or geographic groups, especially in terms of recall for the Detractor class.


# 11. Conclusion

This Business Understanding phase sets the foundation of the project. The company is facing a significant lack of visibility on customer satisfaction, since only **15% of the customer base** is covered by a direct NPS measurement.

The main challenge is twofold: on one hand, better identifying at-risk customers in order to prioritize retention actions, and on the other hand, extending the understanding of customer satisfaction to the entire customer base using a predictive model.

Translating this challenge into a machine learning problem already requires several key decisions. We must consider the **ordinal nature** of the target, the **data leakage risk** linked to variables such as Churn Score and Churn Value, as well as the **class imbalance** problem. We also need to anticipate issues related to **generalization** and **fairness across customer segments**.

These elements will guide all the next steps of the project, from data preparation to modeling and evaluation.
